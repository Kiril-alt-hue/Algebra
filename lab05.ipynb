### 1. Абелеві групи




1. Напишіть функцію, яка для заданого $n$ повертає кількість попарно неізоморфних абелевих груп порядку $n$.
2. Знайдіть перші сто значень $n$, для яких існує рівно одна абелева група порядку $n$, і порівняйте з розкладом $n$ у добуток простих.
3. Напишіть гіпотезу про те, для яких $n$ існує рівно одна абелева група порядку $n$.
4. Спробуйте знайти всі $n$, для яких існує рівно $n$ абелевих груп порядку $n$.

```python
# --- Завдання 1: Функція для підрахунку груп ---
print("--- 1. Функція для підрахунку неізоморфних абелевих груп ---")

def count_abelian_groups(n):
  
    if n <= 0:
        return 0
    if n == 1:
        return 1
    
    total_count = 1
    
    prime_factors = factor(n) 
    
    for p, k in prime_factors:
        # Кількість груп порядку p^k дорівнює кількості розбиттів P(k)
        total_count *= Partitions(k).cardinality()
        
    return total_count

# Тестування функції:
print(f"Кількість груп порядку 1: {count_abelian_groups(1)}")
print(f"Кількість груп порядку 12 (2^2 * 3^1 -> P(2)*P(1) = 2*1): {count_abelian_groups(12)}")
print(f"Кількість груп порядку 100 (2^2 * 5^2 -> P(2)*P(2) = 2*2): {count_abelian_groups(100)}")
print(f"Кількість груп порядку 1024 (2^10 -> P(10)): {count_abelian_groups(1024)}")


# --- Завдання 2: n, для яких існує 1 група ---
print("\n--- 2. Перші 100 значень n, для яких існує 1 абелева група ---")

square_free_numbers = []
n = 1
while len(square_free_numbers) < 100:
    if count_abelian_groups(n) == 1:
        square_free_numbers.append(n)
    n += 1

print(square_free_numbers)

print("\nПорівняння з розкладом на прості множники (перші 15):")
for i in range(15):
    n_val = square_free_numbers[i]
    print(f"n = {n_val:<3} | Розклад: {factor(n_val)}")


# --- Завдання 3: Гіпотеза ---
print("\n--- 3. Гіпотеза ---")
print("Рівно одна попарно неізоморфна абелева група порядку n існує")
print("тоді і тільки тоді, коли n є 'безквадратним числом' (square-free).")
print("Тобто, у розкладі n на прості множники (n = p1^k1 * ... * pr^kr)")
print("всі показники степенів k_i дорівнюють 1.")
print("Це тому, що P(k) = 1 (кількість розбиттів) тільки для k=1.")


# --- Завдання 4: Пошук n, для яких Count(n) = n ---
print("\n--- 4. Пошук n, для яких кількість груп дорівнює n ---")

solutions_task4 = []
# Перевіримо перші 1 000 000 значень
search_limit = 1000000
for n_val in range(1, search_limit + 1):
    if count_abelian_groups(n_val) == n_val:
        solutions_task4.append(n_val)

print(f"Знайдені розв'язки (до n = {search_limit}): {solutions_task4}")
print("\nТеоретичне обґрунтування:")
print("Count(n) = П P(k_i) і n = П p_i^(k_i).")
print("Оскільки p_i^(k_i) росте експоненційно, а P(k_i) росте набагато повільніше,")
print(f"і p_i^(k_i) > P(k_i) для всіх n > 1, то єдиним розв'язком є n=1.")
```

### 2. Дільники нуля в кільцях

1. Задайте кільця $\mathbb{Z}_{30}$, $\mathbb{Z}_{12}\times\mathbb{Z}_{15}$, $M_3(\mathbb{Z}_3)$, $\mathbb{Z}_6[x]$.
2. Знайдіть кількість дільники нуля в цих кільцях.
3. Знайдіть мультиплікативну групу та її порядок. 



```python
# --- 1. Кільце Z_30 ---
print("--- 1. Кільце R = Z_30 ---")
R_30 = Integers(30)
R_30_size = R_30.cardinality()

# Мультиплікативна група R* (одиниці)
R_30_units_order = R_30.unit_group().order()

# Кількість дільників нуля = |R| - |R*| - 1 (за 0)
R_30_zero_divisors = R_30_size - R_30_units_order - 1

print(f"|R|: {R_30_size}")
print(f"Мультиплікативна група |R*| (порядок): {R_30_units_order}")
print(f"Кількість дільників нуля: {R_30_zero_divisors}")


# --- 2. Кільце Z_12 x Z_15 ---
print("\n--- 2. Кільце R = Z_12 x Z_15 ---")
R_12 = Integers(12)
R_15 = Integers(15)

# Розмір добутку |R| = |R_12| * |R_15|
R_prod_size = R_12.cardinality() * R_15.cardinality()

# Порядок групи одиниць добутку |R*| = |R_12*| * |R_15*|
R_prod_units_order = R_12.unit_group().order() * R_15.unit_group().order()

# Кількість дільників нуля = |R| - |R*| - 1
R_prod_zero_divisors = R_prod_size - R_prod_units_order - 1

print(f"|R|: {R_prod_size}")
print(f"Мультиплікативна група |R*| (порядок): {R_prod_units_order}")
print(f"Кількість дільників нуля: {R_prod_zero_divisors}")


# --- 3. Кільце M_3(Z_3) ---
print("\n--- 3. Кільце R = M_3(Z_3) ---")
# Z_3 - це скінченне поле, GF(3)
K = GF(3) 
R_M3 = MatrixSpace(K, 3, 3) # Кільце матриць 3x3 над K

R_M3_size = R_M3.cardinality()

# Мультиплікативна група R* - це Загальна Лінійна Група GL(n, K)
R_M3_units = GL(3, K)
R_M3_units_order = R_M3_units.order()

R_M3_zero_divisors = R_M3_size - R_M3_units_order - 1

print(f"|R|: {R_M3_size} (тобто 3^9)")
print(f"Мультиплікативна група |R*| (порядок): {R_M3_units_order}")
print(f"Кількість дільників нуля: {R_M3_zero_divisors}")


# --- 4. Кільце Z_6[x] ---
print("\n--- 4. Кільце R = Z_6[x] ---")
R_6 = Integers(6)
# R_poly - кільце многочленів
R_poly = PolynomialRing(R_6, 'x') 
P = R_poly

print(f"Чи є кільце скінченним? {P.is_finite()}")
print(f"Загальний розмір |R|: {P.cardinality()}")

# Мультиплікативна група для R[x]
R_poly_units_order = 2
print(f"Мультиплікативна група |R*| (порядок): {R_poly_units_order} (це {R_6.unit_group()})")

# Дільники нуля
print("Кількість дільників нуля: Нескінченна")
print("Приклади дільників нуля:")

# Отримаємо 'x' як змінну
x = P.gen() 

print(f"  2 * 3 = {P(2) * P(3)}") 
print(f"  3 * 2x^2 = {P(3) * (2*x^2)}")
print(f"  4 * 3x = {P(4) * (3*x)}")
print("  Нескінченна родина (3 * 2x^n = 0):")
print(f"  3 * 2*x^5 = {P(3) * (2*x^5)}")
```

### 3. Корінь з одиниці за модулем $p$

1. Знайдіть перші сто простих чисел $p$, для яких факторкільце $\mathbb{Z}_p[x]/(x^2+1)$ є полем.  $\qquad$ % PolynomialRing(GF(p))
2. Знайдіть остачі при ділення цих простих чисел на $m$ для деяких значень $m=2,3,\ldots,10$.        
3. Спробуйте знайти закономірність і сформулювати гіпотезу про такі прості числа.


```python
# --- Завдання 1: Пошук перших 100 простих чисел ---

field_primes = []
p = 2 # Почнемо з першого простого числа

print("--- 1. Пошук перших 100 простих чисел p ---")

while len(field_primes) < 100:
    
    # Створюємо поле Z_p
    K = GF(p) 
    # Створюємо кільце Z_p[x]
    R, x = PolynomialRing(K, 'x').objgen() 
    f = x^2 + 1
    
    # Перевіряємо, чи є многочлен незвідним
    if f.is_irreducible():
        field_primes.append(p)
        
    p = next_prime(p) 

print(f"Знайдено {len(field_primes)} простих чисел, для яких Z_p[x]/(x^2+1) є полем:")
print(field_primes)


# --- Завдання 2: Аналіз остач ---
print("\n--- 2. Аналіз остач для знайдених простих чисел ---")

for m in range(2, 11):
    # Використовуємо 'set' для пошуку унікальних остач
    remainders = {pr % m for pr in field_primes}
    
    print(f"Унікальні остачі при діленні на m = {m}: {sorted(list(remainders))}")


# --- Завдання 3: Формулювання гіпотези ---
print("\n--- 3. Гіпотеза ---")
print("Аналізуючи вивід Завдання 2, ми бачимо чітку закономірність:")
print("Унікальні остачі при діленні на m = 4: [3]")
print("\nЦе приводить нас до наступної гіпотези:")
print("Кільце Z_p[x]/(x^2+1) є полем тоді і тільки тоді,")
print("коли p є простим числом, що задовольняє умову p = 3 (mod 4).")
```

### 4. Ідеали та факторкільця кільця многочленів

1. Задайте кільце многочленів $R=\mathbb{Z}_{7}[x]$. $\qquad$  %  R.\<x\>=Integers(7)[]
3. Задайте ідеали $I_1=(2x^3+5x+1)$, $I_2=(x^4+3x^3+2)$, $I_3=(x^5+x^3+x+1)$.  $\qquad$    %  I = R.ideal(f(x))
4. Побудуйте факторкільця $R/I$ і знайдіть його порядок. $\qquad$  %  R.quotient(І)
5. Перевірте, для яких ідеалів факторкільце $R/I$ є полем.
6. Перевірте, чи має елемент $(x^5+3x^2+2)+I$ обернений у факторкільці $R/I$ та знайдіть обернений.

```python
# 1. Визначити кільце R та змінну x
R, x = PolynomialRing(GF(7), 'x').objgen()
print(f"Базове кільце: R = {R}")

# 2. Визначити многочлени для ідеалів
f1 = 2*x^3 + 5*x + 1
f2 = x^4 + 3*x^3 + 2
f3 = x^5 + x^3 + x + 1

# 3. Визначити елемент g
g_poly = x^5 + 3*x^2 + 2
print(f"Елемент для перевірки: g = {g_poly}")

# 4. Створити список для ітерації
polynomials = [f1, f2, f3]

# 5. Обробити кожен ідеал
for i, f in enumerate(polynomials, 1):
    print(f"\n{'='*40}")
    print(f"--- АНАЛІЗ ДЛЯ I_{i} = ({f}) ---")
    
    # Створити ідеал
    I = R.ideal(f)
    
    # Побудувати факторкільце
    Q = R.quotient(I)
    print(f"Факторкільце Q_{i} = R/I_{i} побудовано.")
    
    # Знайти порядок
    # Порядок |R/(f)| = |K|^deg(f)
    order = Q.cardinality()
    print(f"Порядок |Q_{i}|: {order} (тобто 7^{f.degree()})")
    
    # Перевірити, чи є полем
    # Q = R/I є полем <=> I є максимальним <=> f є незвідним
    is_field = f.is_irreducible()
    print(f"Чи є Q_{i} полем? {is_field}")
    if not is_field:
        # Якщо ні, покажемо чому (знайдемо факторизацію)
        print(f"  Причина: многочлен f_{i} є звідним. Факторизація: {f.factor()}")
        
    print(f"\n--- Перевірка елемента g = ({g_poly}) + I_{i} ---")
    
    # Отримати елемент g в Q 
    g_element = Q(g_poly)
    print(f"Елемент g в кільці Q_{i} (після редукції): {g_element}")

    # Перевірити на оберненість
    # Елемент g+I є оберненим в R/(f) <=> НСД(f, g) = 1 (або константа)
    common_divisor = gcd(f, g_poly)
    # is_unit() перевіряє, чи є НСД константою (не 0)
    is_invertible = common_divisor.is_unit()
    
    print(f"Чи має g обернений елемент? {is_invertible}")
    
    
    #xbar - це х з рискою
    
    
    if is_invertible:
        g_inverse = 1 / g_element
        print(f"Обернений елемент g^-1: {g_inverse}")
        
        # Перевірка (можна розкоментувати)
        # print(f"Перевірка: g * g^-1 = {g_element * g_inverse}")
    else:
        print(f"Елемент g не має оберненого.")
        print(f"  Причина: НСД(f_{i}, g) = {common_divisor}, що не є одиницею.")
```
