### 1. Приклади груп

Для кожної з груп $G$= $S_3$, $D_4$, $\mathbb{Z}_{14}$, $\mathbb{Z}_{14}^{*}$ виконайте наступні кроки:
1. Створіть групу.
2. Знайдіть порядок групи.     # G.order()
3. Побудуйте таблицю Келі групи.    # G.cayley_table(names='elements')
4. Виберіть довільні два елементи в групі.
5. Виконайте групову операцію над обраними елементами.
6. Знайдіть обернені для обраних елементів.
7. Визначте, чи є група абелевою. Якщо ні, то знайдіть два елементи групи, які не комутують.   # G.is_abelian()

```python
print("=== Група S3 ===")
S3 = SymmetricGroup(3)


print("Елементи групи:", list(S3))


print("Порядок групи:", S3.order())


print("Таблиця Келі:")
display(S3.cayley_table(names='elements'))


elements_S3 = list(S3)
e1_S3, e2_S3 = elements_S3[1], elements_S3[2]
print(f"Вибрані елементи: {e1_S3}, {e2_S3}")


op_S3 = e1_S3 * e2_S3
print(f"Операція над вибраними елементами: {e1_S3} * {e2_S3} = {op_S3}")


inv1_S3 = e1_S3^-1
inv2_S3 = e2_S3^-1
print(f"Обернені елементи: {e1_S3} -> {inv1_S3}, {e2_S3} -> {inv2_S3}")


is_abelian_S3 = S3.is_abelian()
print("Чи є група абелевою?:", is_abelian_S3)
if not is_abelian_S3:
    # знайдемо два елементи, які не комутують
    elements = list(S3)
    found = False
    for a in elements:
        for b in elements:
            if a * b != b * a:
                print(f"Приклад двох елементів, які не комутують: {a}, {b}")
                found = True
                break
        if found:
            break

print("\n" + "="*50 + "\n")
```

### 3. Підгрупи і гратка підгруп

Для кожної з груп $\mathbb{Z}_{24}$, $Q_8$, $D_5$, $S_4$ виконайте наступні кроки:
1. Створіть групу.
2. Знайдіть усі її підгрупи.   % G.subgroups()
3. Побудуйте гратку підгруп.   % G.subgroup_lattice() може не працювати, запитайте Gemini or ChatGpt

```python

Z24 = CyclicPermutationGroup(24)
print(f"Група створена: {Z24}")
print(f"Порядок групи: {Z24.order()}")
print(f"Абелева? {Z24.is_abelian()}\n")


subgroups_Z24 = Z24.subgroups()
print(f"Кількість підгруп: {len(subgroups_Z24)}")
print("Список підгруп (порядок):")
for i, sg in enumerate(subgroups_Z24):
    print(f"  Підгрупа {i+1}: порядок = {sg.order()}")
print()


print("Гратка підгруп (порядки підгруп):")
orders = sorted(set(sg.order() for sg in subgroups_Z24))
for order in orders:
    subgr_of_order = [sg for sg in subgroups_Z24 if sg.order() == order]
    print(f"Підгрупи порядку {order}: {len(subgr_of_order)} штук")
print("\n" + "="*50 + "\n")
```

```python
print("=== ГРУПА Q8 ===")
Q8 = QuaternionGroup()
print(f"Група створена: {Q8}")
print(f"Порядок групи: {Q8.order()}")
print(f"Абелева? {Q8.is_abelian()}\n")


subgroups_Q8 = Q8.subgroups()
print(f"Кількість підгруп: {len(subgroups_Q8)}")
print("Список підгруп (порядок та генератори):")
for i, sg in enumerate(subgroups_Q8):
    print(f"  Підгрупа {i+1}: порядок = {sg.order()}, генератори = {sg.gens()}")
print()


print("Гратка підгруп:")
print("1 (тривіальна)")
print("  ↳ 2 ({-1, 1})")
print("    ↳ 4 (циклічні підгрупи <i>, <j>, <k>)")
print("      ↳ 8 (сама Q8)")
print("\n" + "="*50 + "\n")


print("=== АНАЛІЗ: ПОШУК ГРУПИ З АБЕЛЕВИМИ ВЛАСНИМИ ПІДГРУПАМИ ===")

groups = {
    "Z24": Z24,
    "Q8": Q8, 
    "D5": D5,
    "S4": S4
}

for name, G in groups.items():
    if not G.is_abelian():
        print(f"\nАналіз групи {name}:")
        all_proper_subgroups_abelian = True
        non_abelian_subgroups = []
        

        for H in G.subgroups():
            if H.order() != 1 and H.order() != G.order():
                if not H.is_abelian():
                    all_proper_subgroups_abelian = False
                    non_abelian_subgroups.append(H.order())
        
        if all_proper_subgroups_abelian:
            print(f"  ✓ {name} - неабелева, але всі власні підгрупи абелеві!")
        else:
            print(f"  ✗ {name} - має неабелеві власні підгрупи порядків: {set(non_abelian_subgroups)}")

print("\n" + "="*50)
```

3. Знайдіть всі мінімальні системи твірних групи.    % G.minimal_generating_sets()
4. Візьміть довільні два елементи групи, породіть ними підгрупу, знайдіть її порядок.  % G.subgroup([el1, el2])

```python
print("=== ГРУПА Z24 ===")
Z24 = CyclicPermutationGroup(24)
print(f"Група створена: {Z24}")
print(f"Порядок групи: {Z24.order()}")
print(f"Абелева? {Z24.is_abelian()}\n")


gens = Z24.gens()
print(f"Система твірних: {gens}")


print("Мінімальні системи твірних (елементи порядку 24):")
minimal_gens = []
for el in Z24:
    if el.order() == 24:
        minimal_gens.append([el])
        print(f"  {[el]}")
print()


el1 = Z24.random_element()
el2 = Z24.random_element()
print(f"Елемент 1: {el1}, порядок: {el1.order()}")
print(f"Елемент 2: {el2}, порядок: {el2.order()}")

subgroup = Z24.subgroup([el1, el2])
print(f"Підгрупа, породжена елементами: порядок = {subgroup.order()}")
print("\n" + "="*50 + "\n")
```

### 5. Графи Келі

Графом Келі групи $G$ відносно системи твірних $S$ називається орієнтований граф $\Gamma(G,S)$, вершинами якого є елементи групи і кожний $g\in G$ з'єднаний орієнтованим ребро з $gs$ для всіх $s\in S$.

1. Побудуйте графи Келі для груп $D_4$, $Q_8$, $S_3$, $S_4$, $A_4$ для деяких систем твірних.  % G.cayley_graph()
2. Опишіть спостереження: які особливості бачите у графів Келі груп?

```python
print("=== ГРАФ КЕЛІ ДЛЯ A4 ===")
A4 = AlternatingGroup(4)
print(f"Порядок групи: {A4.order()}")


S = A4.gens()
print(f"Система твірних: {S}")


cg = A4.cayley_graph(generators=S)
print(f"Граф Келі має {cg.num_verts()} вершин та {cg.num_edges()} ребер")


plot = cg.plot(color_by_label=True, vertex_size=500, vertex_labels=True)
plot.show()
print("\n" + "="*50 + "\n")


print("=== АНАЛІЗ ОСОБЛИВОСТЕЙ ГРАФІВ КЕЛІ ===")

groups = {
    "D4": DihedralGroup(4),
    "Q8": QuaternionGroup(),
    "S3": SymmetricGroup(3),
    "S4": SymmetricGroup(4),
    "A4": AlternatingGroup(4)
}

observations = []

for name, G in groups.items():
    S = G.gens()
    cg = G.cayley_graph(generators=S)
    
    degree_info = "N/A"
    if cg.num_verts() > 0:
        degrees = set(cg.degree())
        degree_info = f"{degrees}-регулярний"
    
    observations.append({
        "Група": name,
        "Порядок": G.order(),
        "Твірних": len(S),
        "Вершин": cg.num_verts(),
        "Ребер": cg.num_edges(),
        "Регулярність": degree_info,
        "Зв'язність": "Зв'язний" if cg.is_connected() else "Не зв'язний"
    })

print("ОСОБЛИВОСТІ ГРАФІВ КЕЛІ:")
print("1. Всі графи Келі є регулярними (кожна вершина має однаковий ступінь)")
print("2. Всі графи Келі є зв'язними (з будь-якої вершини можна дійти до будь-якої іншої)")
print("3. Графи мають симетрію, що відображає структуру групи")
print("4. Кількість ребер, що виходять з кожної вершини, дорівнює кількості твірних")
print("5. Графи для абелевих груп мають більш регулярну структуру")
print("6. Графи для неабелевих груп мають більш складну структуру")

print("\nТАБЛИЦЯ ВЛАСТИВОСТЕЙ:")
for obs in observations:
    result_str = (f"""{obs['Група']}: порядок {obs['Порядок']}, {obs['Твірних']} твірних, 
{obs['Вершин']} вершин, {obs['Ребер']} ребер, 
{obs['Регулярність']}, {obs["Зв'язність"]}""")
    print(result_str)
```
