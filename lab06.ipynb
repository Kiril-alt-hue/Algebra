### 1. Розширення полів

1. Побудуйте розширення $K=\mathbb{Q}(\sqrt[3]{2},i)$ поля $\mathbb{Q}$.
2. Знайдіть степінь розширення $[K:\mathbb{Q}]$ та базис $K$ над $\mathbb{Q}$.
3. Задайте елемент $a=\sqrt[3]{2}+i$ і знайдіть його мінімальний многочлен $m_a(x)$ над полем $\mathbb{Q}$.
4. Побудуйте розширення $L=\mathbb{Q}(\sqrt[3]{2}+i)$ як факторкільце $\mathbb{Q}[x]/(m_a(x))$.
5. Перевірте, чи є поля $K$ та $L$ рівними, порівнявши степені розширення.



```python
# 1. Побудова розширення K = Q(cbrt(2), i)
R.<x> = QQ[]
# Створюємо поле K
K.<cbrt2, i_gen> = NumberField([x^3 - 2, x^2 + 1])

print("--- Завдання 1 ---")
print(f"Поле K побудовано: {K}")

# 2. Степінь розширення та базис K над Q
degree_K = K.absolute_degree()
# Вручну будуємо базис як всі можливі добутки степенів генераторів
basis_K = [cbrt2^j * i_gen^k for j in range(3) for k in range(2)]

print("\n--- Завдання 2 ---")
print(f"Степінь [K : Q] = {degree_K}")
print("Базис K над Q:")
print(basis_K)

# 3. Елемент a = cbrt(2) + i та його мінімальний многочлен
a_elem = cbrt2 + i_gen

# Використовуємо absolute_minpoly(), 
# щоб отримати многочлен над Q (без i в коефіцієнтах)
min_poly_a = a_elem.absolute_minpoly()

print("\n--- Завдання 3 ---")
print(f"Елемент a = {a_elem}")
print(f"Мінімальний многочлен m_a(x) (над Q) = {min_poly_a}")

# 4. Побудова розширення L = Q(a)
L.<alpha> = NumberField(min_poly_a)

print("\n--- Завдання 4 ---")
print(f"Поле L побудовано: {L}")

# 5. Перевірка рівності
degree_L = L.absolute_degree()

print("\n--- Завдання 5 ---")
print(f"Степінь [K : Q] = {degree_K}")
print(f"Степінь [L : Q] = {degree_L}")

if degree_K == degree_L:
    print("Висновок: Поля K та L рівні (степені збігаються).")
else:
    print("Висновок: Поля різні.")
```

### 2. Поля розкладу

1. Побудуйте поле розкладу $L$ многочлена $f(x)=x^4+x^2+1$ над полем $\mathbb{Q}$.
2. Знайдіть степінь та базис $L$ над $\mathbb{Q}$.
3. Розкладіть многочлен $f(x)$ на лінійні множники над полем $L$.
4. Побудуйте поле розкладу $K$ многочлена $g(x)=x^5+3x^2+2$ над полем $\mathbb{Z}_3$.
5. Знайдіть степінь та базис $K$ над $\mathbb{Z}_3$.
6. Розкладіть многочлен $g(x)$ на лінійні множники над полем $K$.

```python
print("--- 1. Поле розкладу L для f(x) = x^4 + x^2 + 1 над Q ---")

# 1. Побудова поля розкладу L
R.<x> = QQ[]
f = x^4 + x^2 + 1
L.<a> = f.splitting_field()

print(f"Многочлен: {f}")
print(f"Поле розкладу L: {L}")

# 2. Степінь та базис L над Q
deg_L = L.degree()

basis_L = [a^i for i in range(deg_L)]

print(f"\n--- 2. Степінь та базис L над Q ---")
print(f"Степінь [L : Q] = {deg_L}")
print(f"Базис L над Q: {basis_L}")

# 3. Розклад многочлена f(x) на лінійні множники над L
# Переходимо в кільце многочленів над полем L
RL.<y> = L[]
f_factors = f.change_ring(L).factor()

print(f"\n--- 3. Розклад f(x) над L ---")
print(f_factors)

print("\n\n--- 4. Поле розкладу K для g(x) = x^5 + 3x^2 + 2 над Z3 ---")

# 4. Побудова поля розкладу K
# Спочатку задаємо базове поле Z3 (GF(3))
F3 = GF(3)
R3.<z> = F3[]
# В Z3 коефіцієнт 3 перетвориться на 0, тому g = z^5 + 2
g = z^5 + 3*z^2 + 2 

# 'b' буде генератором розширеного скінченного поля
K.<b> = g.splitting_field()

print(f"Многочлен (над Z3): {g}")
print(f"Поле розкладу K: {K}")

# 5. Степінь та базис K над Z3
deg_K = K.degree()
basis_K = [b^i for i in range(deg_K)]

print(f"\n--- 5. Степінь та базис K над Z3 ---")
print(f"Степінь [K : Z3] = {deg_K}")
print(f"Базис K над Z3: {basis_K}")

# 6. Розклад многочлена g(x) на лінійні множники над K
RK.<w> = K[]
g_factors = g.change_ring(K).factor()

print(f"\n--- 6. Розклад g(x) над K ---")
print(g_factors)
```

### 3. Скінченні поля

1. Побудуйте поле $F$ з $8$ елементів. Виведіть таблицю множення.
2. Побудуйте факторкільця $K=\mathbb{Z}_2[x]/(x^3+x+1)$ та $L=\mathbb{Z}_2[x]/(x^3+x^2+1)$.
3. Перевірте, що $K$ та $L$ є полями з $8$ елементів.
4. Знайдіть первісні елементи для полів $K$ та $L$.
5. Встановіть ізоморфізм між полями $K$ та $L$.


```python
# --- 1. Побудова поля F з 8 елементів та таблиця множення ---
print("--- 1. Поле F (8 елементів) ---")
F = GF(8, 'a') 
print(f"Поле F: {F}")
print("Таблиця множення поля F:")
# Виводимо таблицю множення (використовуємо імена елементів)
print(F.multiplication_table(names='elements'))


# --- 2. Побудова фактор-кілець K та L ---
print("\n--- 2. Побудова фактор-кілець K та L ---")
# Кільце многочленів над Z2
R.<x> = GF(2)[]

# Визначаємо ідеали (многочлени)
poly_K = x^3 + x + 1
poly_L = x^3 + x^2 + 1

# Створюємо фактор-кільця
# 'u' буде генератором для K (тобто класом x)
# 'v' буде генератором для L
K = R.quotient(poly_K, names='u')
L = R.quotient(poly_L, names='v')

print(f"K = Z2[x] / ({poly_K})")
print(f"L = Z2[x] / ({poly_L})")


# --- 3. Перевірка властивостей полів ---
print("\n--- 3. Перевірка, чи є K та L полями з 8 елементів ---")
print(f"Чи є K полем? {K.is_field()}")
print(f"Кількість елементів K: {K.cardinality()}")
print(f"Чи є L полем? {L.is_field()}")
print(f"Кількість елементів L: {L.cardinality()}")


# --- 4. Пошук первісних елементів ---
print("\n--- 4. Первісні елементи ---")

# Елемент g первісний, якщо його степені генерують всі 7 ненульових елементів
def find_primitives(field, name):
    prims = []
    size = field.cardinality()
    for elem in field:
        if elem == 0: continue
        # Перевіряємо порядок елемента
        # У Sage для generic quotient ring метод multiplicative_order може бути не прямим,
        # тому перевіримо вручну: чи g^(size-1) == 1 і для менших степенів != 1
        is_prim = True
        # Перевіряємо дільники порядку групи (для 7 це тільки 1)
        # Але простіше: чи elem^i != 1 для i від 1 до 6
        current = elem
        seen = {current}
        for _ in range(size - 2): # генеруємо решту
            current *= elem
            seen.add(current)
        
        if len(seen) == size - 1: # якщо згенерували всі 7 ненульових
            prims.append(elem)
    return prims

prims_K = find_primitives(K, "K")
prims_L = find_primitives(L, "L")

print(f"Первісні елементи в K: {prims_K}")
print(f"Первісні елементи в L: {prims_L}")


# --- 5. Встановлення ізоморфізму ---
print("\n--- 5. Ізоморфізм між K та L ---")
# Щоб побудувати ізоморфізм phi: K -> L, нам треба перевести генератор u (з K)
# в такий елемент y (в L), який є коренем многочлена, що визначає K.
# Тобто шукаємо y в L таке, що y^3 + y + 1 = 0.

u = K.gen()
root_in_L = None

for elem in L:
    if elem^3 + elem + 1 == 0:
        root_in_L = elem
        # Ми беремо перший знайдений корінь
        break

print(f"Генератор u з K (корінь x^3+x+1) має перейти в елемент {root_in_L} з поля L")

phi = K.hom([root_in_L])
print("Ізоморфізм phi побудовано.")

# Перевірка (приклад)
val = u^2 + 1
mapped_val = phi(val)
print(f"Приклад відображення: phi({val}) = {mapped_val}")
```
